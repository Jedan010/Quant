#### Square-Root Diffusion

$$ dx_t = \kappa(\theta - x_t)dt + \sigma\sqrt{x_t}dZ_t $$

```python
x0 = 0.05
kappa = 3.0
theta = 0.02
sigma = 0.1
```

```python
I = 10000
M = 50
dt = T / M
```

```python
def srd_euler():
    xh = np.zeros((M+1, I))
    x1 = np.zeros_like(xh)
    xh[0] = x0
    x1[0] = x0
    for t in range(1, M+1):
        xh[t] = (xh[t-1] + kappa*(theta-np.maximum(xh[t-1],0))*dt + sigma*np.sqrt(np.maximum(xh[t-1], 0))*np.sqrt(dt)*npr.randn(I))
    x1 = np.maximum(xh, 0)
    return x1
```

```python
x1 = srd_euler()
```

```python
plt.hist(x1[-1], bins=50);
```

```python
plt.plot(x1[:, 5:10]);
```

$$ x_t = \frac{\sigma^2(1-e^{-\kappa\Delta t})}{4\kappa}\chi_{d}^{'2}(\frac{4\kappa e^{-\kappa\Delta t}}{\sigma^2(1-e^{-\kappa\Delta t})}x_s) $$

```python
def srd_exact():
    x2 = np.zeros((M+1, I))
    x2[0] = x0
    for t in range(1, M+1):
        df = 4 * theta * kappa / sigma**2
        c = (sigma**2 * (1-np.exp(-kappa*dt))) / (4*kappa)
        nc = np.exp(-kappa*dt) / c*x2[t-1]
        x2[t] = c*npr.noncentral_chisquare(df, nc, size=I)
    return x2
x2 = srd_exact()
```

```python
plt.hist(x2[-1], bins=50);
```

### Stochastic Volatility

$$ dS_t = rS_tdt + \sqrt{v_t}S_tdZ^1_t $$
$$ dv_t = \kappa(\theta - v_t)dt + \sigma\sqrt{v_t}dZ^2_{t} $$
$$ dZ^1_tdZ^2_t = \rho $$

```python
S0 = 100.
r = 0.05
v0 = 0.1
kappa = 3.0
theta = 0.25
sigma = 0.1
rho = 0.6
T = 1.0
```

$$ \Sigma = LL^T $$
$$ Z = LX  $$
$$ E[ZZ^T] = E[(LX)(LX)^T]= E[LXX^TL^T]=LE[XX^T]L^T=LIL^T=\Sigma$$

```python
corr_mat = np.array([[1,rho], [rho, 1]])

cho_mat = np.linalg.cholesky(corr_mat)
```

```python
cho_mat
```

```python
M = 50
I = 10000
dt = T / M
ran_num = npr.standard_normal((2, M + 1, I))
```

```python
v = np.zeros((M+1, I))
v[0] = v0
for t in range(1, M+1):
    Z = np.dot(cho_mat, ran_num[:,t,:])
    v[t] = v[t-1] + kappa*(theta-np.maximum(v[t-1],0))*dt + sigma*np.sqrt(np.maximum(v[t-1], 0))*Z[1]*np.sqrt(dt)
v = np.maximum(v, 0)
```

```python
S = np.zeros((M+1, I))
S[0] = S0
for t in range(1, M+1):
    Z = np.dot(cho_mat, ran_num[:,t,:])
    S[t] = S[t-1] + r*S[t-1]*dt + np.sqrt(v[t])*S[t-1]*Z[0]*np.sqrt(dt)
```

### Jump-Diffusion

$$ dS_t = (r-r_J)S_tdt + S_tdZ_t + J_tS_tdN_t $$
$$ Z_t \sim N(0,t), \space N_t \sim P(\lambda t) $$

```python
S0 = 100.
r = 0.05
sigma = 0.2
lamb = 0.75
mu = -0.6
delta = 0.25
T = 1.0
```

```python
M = 50
I = 10000
dt = T / M
rj = lamb * (np.exp(mu + 0.5 * delta ** 2) - 1)
S = np.zeros((M + 1, I))
S[0] = S0
sn1 = npr.standard_normal((M + 1, I))
sn2 = npr.standard_normal((M + 1, I))
poi = npr.poisson(lamb * dt, (M + 1, I))
for t in range(1, M + 1, 1):
    S[t] = S[t - 1] * (np.exp((r - rj - 0.5 * sigma ** 2) * dt
                       + sigma * np.sqrt(dt) * sn1[t])
                       + (np.exp(mu + delta * sn2[t]) - 1)
                       * poi[t])
    S[t] = np.maximum(S[t], 0)
```

## Variance Reduction

```python
def gen_sn(M, I, anti_paths=True, mo_match=True):
    ''' Function to generate random numbers for simulation.
    
    Parameters
    ==========
    M : int
        number of time intervals for discretization
    I : int
        number of paths to be simulated
    anti_paths: boolean
        use of antithetic variates
    mo_math : boolean
        use of moment matching
    '''
    if anti_paths is True:
        sn = npr.standard_normal((M , int(I / 2)))
        sn = np.concatenate((sn, -sn), axis=1)
    else:
        sn = npr.standard_normal((M , I))
    if mo_match is True:
        sn = (sn - sn.mean()) / sn.std()
    return sn
```

## Valuation

### European Options

$$ C = e^{-rT}E[h(S_T)]=e^{-rT}\int_{-\infty}^{\infty}h(s)f(s)ds $$
$$ C = e^{-rT}\frac{1}{N}\sum_{i=1}^{N}h(S_T^i) $$

```python
S0 = 100.
r = 0.05
sigma = 0.25
T = 1.0
I = 50000
```

```python
sn = gen_sn(1, I)
```

```python
def gbm_mcs_stat(K):
    sn = gen_sn(1, I)
    ST = S0 * np.exp((r-1/2*sigma**2)*T + sigma*T**0.5*sn[0])
    hT = np.maximum(ST - K, 0)
    C = np.exp(-r*T) * np.mean(hT)
    return C
```

```python
gbm_mcs_stat(105)
```

```python
M = 50
dt = T/M
def gbm_mcs_dyna(K):
    S = np.zeros((M+1, I))
    S[0] = S0
    sn = gen_sn(M, I)
    for t in range(1, M+1):
        S[t] = S[t-1] + r*S[t-1]*dt + sigma*S[t-1]*dt**0.5*sn[t-1]
    hT = np.maximum(S[-1] - K, 0)
    C = np.exp(-r*T) * np.mean(hT)
    return C
```

```python
gbm_mcs_dyna(105)
```

### American Options

```python
def gbm_mcs_amer(K, option='call'):
    ''' Valuation of American option in Black-Scholes-Merton
    by Monte Carlo simulation by LSM algorithm
    
    Parameters
    ==========
    K : float
        (positive) strike price of the option
    option : string
        type of the option to be valued ('call', 'put')
    
    Returns
    =======
    C0 : float
        estimated present value of European call option
    '''
    dt = T / M
    df = np.exp(-r * dt)
    # simulation of index levels
    S = np.zeros((M + 1, I))
    S[0] = S0
    sn = gen_sn(M, I)
    for t in range(1, M + 1):
        S[t] = S[t - 1] * np.exp((r - 0.5 * sigma ** 2) * dt 
                + sigma * np.sqrt(dt) * sn[t-1])
    # case based calculation of payoff
    if option == 'call':
        h = np.maximum(S - K, 0)
    else:
        h = np.maximum(K - S, 0)
    # LSM algorithm
    V = np.copy(h)
    for t in range(M - 1, 0, -1):
        reg = np.polyfit(S[t], V[t + 1] * df, 7)
        C = np.polyval(reg, S[t])
        V[t] = np.where(C > h[t], V[t + 1] * df, h[t])
    # MCS estimator
    C0 = df * 1 / I * np.sum(V[1])
    return C0
```

```python
gbm_mcs_amer(110., option='call')
```

## Risk Measures

```python
S0 = 100
r = 0.05
sigma = 0.25
T = 30 / 365.
I = 10000
ST = S0 * np.exp((r - 0.5 * sigma ** 2) * T + sigma * np.sqrt(T) * npr.standard_normal(I))
```

```python
R_gbm = np.sort(ST-S0)
```

```python
plt.hist(R_gbm,bins=50);
```

```python
import scipy.stats as scs
```

```python
percs = [0.01, 0.1, 1., 2.5, 5.0, 10.0]
var = scs.scoreatpercentile(R_gbm, percs)
print("%16s %16s" % ('Confidence Level', 'Value-at-Risk'))
print(33 * "-")
for pair in zip(percs, var):
    print("%16.2f %16.3f" % (100 - pair[0], -pair[1]))
```
