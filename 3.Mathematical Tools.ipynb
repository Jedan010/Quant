## Approximation

```python
def f(x):
    return np.sin(x) + 0.5 * x
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 50)
```

```python
plt.plot(x, f(x), 'b')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
```

### Regression

#### Monomials as Basis Functions

```python
reg = np.polyfit(x, f(x), deg=1)
ry = np.polyval(reg, x)
```

```python
plt.plot(x, f(x),'b',  label='f(x)')
plt.plot(x, ry, 'r.', label='regression')
plt.legend(loc=0)
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
reg = np.polyfit(x, f(x), deg=5)
ry = np.polyval(reg, x)

plt.plot(x,f(x),'b',label='f(x)')
plt.plot(x, ry, 'r.', label='regression')
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
reg = np.polyfit(x, f(x), deg=7)
ry = np.polyval(reg, x)

plt.plot(x,f(x),'b',label='f(x)')
plt.plot(x, ry, 'r.', label='regression')
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
```

## Convex Optimization

```python
def fm(p):
    x, y = p
    return (np.sin(x) + 0.05 * x ** 2
          + np.sin(y) + 0.05 * y ** 2)
```

```python
x = np.linspace(-10, 10, 50)
y = np.linspace(-10, 10, 50)
X, Y = np.meshgrid(x, y)
Z = fm((X, Y))
```

```python
import scipy.optimize as spo
```

```python
spo.brute(fm, ((-10, 10.1, 5), (-10, 10.1, 5)), finish=None)
```

```python
opt1 = spo.brute(fm, ((-10, 10.1, 0.1), (-10, 10.1, 0.1)), finish=None)
opt1
```
