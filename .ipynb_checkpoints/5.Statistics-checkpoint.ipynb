## Normality Tests

### Benchmark Case

```python
def Generate_Geomatric_Brownian_Movement_Path(S0, r, sigma, T, M, I):
    dt = T / M
    S = np.zeros((M + 1, I))
    S[0] = S0
    for t in range(1, M + 1):
        Z = np.random.randn(I)
        Z = (Z - Z.mean()) / Z.std()
        S[t] = S[t-1]*(1 + r*dt + sigma*dt**0.5*Z)
    return S
```

```python
S0 = 100.
r = 0.05
sigma = 0.2
T = 1.0
M = 50
I = 250000
```

```python
S = Generate_Geomatric_Brownian_Movement_Path(S0, r, sigma, T, M, I)
```

```python
plt.plot(S[:,10:20]);
```

```python
log_return = np.log(S[1:]/S[:-1])
```

```python
plt.hist(log_return.flatten(),100, density=True, label='frequency')
x = np.linspace(plt.axis()[0], plt.axis()[1], 100)
plt.plot(x, scs.norm.pdf(x, loc=r / M, scale=sigma / np.sqrt(M)), 'r', label='pdf')
plt.legend()
```

```python
sm.qqplot(log_return.flatten()[::500], line='s');
```

```python
def normality_tests(arr):
    ''' Tests for normality distribution of given data set.
    
    Parameters
    ==========
    array: ndarray
        object to generate statistics on
    '''
    print("Skew of data set  %14.3f" % scs.skew(arr))
    print("Skew test p-value %14.3f" % scs.skewtest(arr)[1])
    print("Kurt of data set  %14.3f" % scs.kurtosis(arr))
    print("Kurt test p-value %14.3f" % scs.kurtosistest(arr)[1])
    print("Norm test p-value %14.3f" % scs.normaltest(arr)[1])
```

## Portfolio Optimization

### The Data

```python
symbols = ['AAPL.O', 'MSFT.O', 'AMZN.O', 'GDX', 'GLD']
noa = len(symbols)
data = raw[symbols]
```

```python
(data / data.iloc[0]*100).plot(figsize=(15,7))
```

```python
ret = np.log(data/data.shift(1))
```

```python
ret.mean() * 252
```

```python
ret.cov() * 252
```

### The Basic Theory

$$ E[r_p]=\mu_p = E[\sum_{i=1}^nw_ir_i]=w^T\mu$$

$$ \sigma_{ij}^2=E[(r_i-\mu_i)(r_j-\mu_j)] $$

$$ \Sigma = (\sigma_{ij})_{i,j}$$

$$var(r_p)=\sigma^2_p=E[(r-\mu)^2]=w^T\Sigma w$$

```python
weights = np.random.random(data.shape[1])
weights /= weights.sum()
weights
```

```python
mu = np.sum(weights * ret.mean())*252
mu
```

```python
sigma2 = np.dot(weights.T, np.dot(ret.cov()*252, weights))
sigma2
```

```python
sigma = np.sqrt(sigma2)
sigma
```

```python
mu_list = []
sigma_list = []
for i in range(3000):
    weights = np.random.random(data.shape[1])
    weights /= weights.sum()
    mu = np.sum(weights * ret.mean())*252
    sigma = np.sqrt(np.dot(weights.T, np.dot(ret.cov()*252, weights)))
    mu_list.append(mu)
    sigma_list.append(sigma)
mu_list = np.array(mu_list)
sigma_list = np.array(sigma_list)
```

```python
plt.figure(figsize=(15,7))
plt.scatter(sigma_list,mu_list, c=(mu_list)/sigma_list)
plt.colorbar(label='Sharpe retio')
plt.xlabel('Expect Volatility $\sigma$')
plt.ylabel('Expect Return r')
```

### Portfolio Optimizations

```python
def statistics(weights):
    weights = np.array(weights)
    mu = np.sum(weights * ret.mean())*252
    sigma = np.sqrt(np.dot(weights.T, np.dot(ret.cov()*252, weights)))
    return np.array([mu, sigma, mu/sigma])
```

```python
import scipy.optimize as opt
```

```python
def min_func_sharpe(weights):
    return -statistics(weights)[2]
```

```python
opts = opt.minimize(fun=min_func_sharpe,
                    x0=np.ones(noa) / noa,
                    method='SLSQP',
                    bounds=[(0, 1)]*noa,
                    constraints=({'type': 'ineq', 'fun': lambda x: 1-np.sum(x)}))
```

```python
opts.x.round(4)
```

```python
statistics(opts.x).round(4)
```

```python
def min_func_var(weights):
    return statistics(weights)[1]**2
```

```python
opts_2 = opt.minimize(fun=min_func_var,
                      x0=np.ones(noa)/noa,
                      method='SLSQP',
                      bounds=[(0, 1)]*noa,
                      constraints=({'type': 'eq', 'fun': lambda x: 1-np.sum(x)}))
```

```python
opts_2.x.round(4)
```

```python
statistics(opts_2.x).round(4)
```

### Efficient Frontier

```python
mu_interval = np.linspace(0,0.25,100)

sigma_interval = np.ones_like(mu_interval)
```

```python
for i, r in enumerate(mu_interval):
    res = opt.minimize(fun=min_func_var,
                       x0=np.ones(noa)/noa,
                       method='SLSQP',
                       bounds=[(0, 1)]*noa,
                       constraints=({'type': 'eq', 'fun': lambda x: 1-np.sum(x)},
                                    {'type': 'eq', 'fun': lambda x: statistics(x)[0]-r}))
    sigma_interval[i] = res.fun**0.5
```

```python
plt.figure(figsize=(12,7))
plt.scatter(sigma_list, mu_list, c=mu_list/sigma_list)
plt.colorbar(label='Sharpe retio')
plt.scatter(sigma_interval, mu_interval,marker='x')
plt.plot(*statistics(opts.x)[0:2][::-1], 'b*', markersize=15)
plt.plot(*statistics(opts_2.x)[0:2][::-1], 'b*', markersize=15)
plt.xlabel('Expect Volatility $\sigma$')
plt.ylabel('Expect Return r')
plt.show()
```

### Capital Market Line

```python
import scipy.interpolate as sci
```

```python
ind = np.argmin(sigma_interval)
erets = mu_interval[ind:]
evols = sigma_interval[ind:]
```

```python
tck = sci.splrep(evols, erets)
```

```python
def f(x):
    ''' Efficient frontier function (splines approximation). '''
    return sci.splev(x, tck, der=0)
def df(x):
    ''' First derivative of efficient frontier function. '''
    return sci.splev(x, tck, der=1)
```

$$ t(x)=a+bx $$
$$ t(0) = r_f \leftrightarrow a = r_f $$
$$ t(x) = f(x) \leftrightarrow a+bx=f(x)$$
$$ t^{'}(x) = f^{'}(x) \leftrightarrow b=f^{'}(x)$$

```python
def equations(p, rf=0.01):
    """
    p[0] = a
    p[1] = b
    p[2] = x
    """
    eq1 = rf - p[0]
    eq2 = rf + p[1]*p[2] - f(p[2])
    eq3 = p[1] - df(p[2])
    return eq1, eq2, eq3
```

```python
fso = opt.fsolve(equations, [0.01, 0.5, 0.15])
```

```python
fso.round(4)
```

```python
plt.figure(figsize=(12,7))
plt.scatter(sigma_list, mu_list, c=mu_list/sigma_list)
plt.colorbar(label='Sharpe retio')
plt.scatter(evols, erets, marker='x')
plt.plot( fso[2],f(fso[2]), 'b*', markersize=15)

cx = np.linspace(0.0, 0.3)
plt.plot(cx, fso[0]+fso[1]*cx, )

# plt.axvline(evols[0], color='b', ls='--', lw=2.0)
# plt.axhline(erets[0], color='b', ls='--', lw=2.0)
plt.axhline(0, color='k', ls='--', lw=2.0)
plt.axvline(0, color='k', ls='--', lw=2.0)

plt.xlabel('Expect Volatility $\sigma$')
plt.ylabel('Expect Return r')
plt.show()
```
